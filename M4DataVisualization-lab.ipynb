## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM MASTER

"""

df = pd.read_sql_query(QUERY,conn)
plt.hist(df, density=True, bins=30, alpha=0.6, color='blue')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM MASTER

"""

df = pd.read_sql_query(QUERY,conn)
sns.boxplot(df, orient='v')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM MASTER

"""

df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(data=df, x='Age', y="WorkWeekHrs")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM MASTER

"""

df = pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', alpha=0.6, sizes=(1,500), color='green')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT *, COUNT(*) AS count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

data = df['count']
labels = df['DatabaseDesireNextYear']

fig, ax = plt.subplots()
ax.pie(data, labels = labels, autopct='%1.1f%%')
plt.show()
```

```python
QUERY = """
SELECT *, COUNT(*) AS count
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

print(df)
```

```python
QUERY = """
SELECT *, COUNT(*) AS count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

data = df['count']
labels = df['LanguageDesireNextYear']

fig, ax = plt.subplots()
ax.pie(data, labels = labels, autopct='%1.1f%%')
plt.show()
```

```python
### Stacked Charts
```

Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM MASTER
WHERE Age BETWEEN 30 AND 35

"""

df = pd.read_sql_query(QUERY,conn)

work = df['WorkWeekHrs'].quantile(0.5)
revision = df['CodeRevHrs'].quantile(0.5)

fig, ax = plt.subplots()

ax.bar('Age 30 - 35', work, label='Work Week Hours')
ax.bar('Age 30 - 35', revision, bottom=work, label='Code Revision Hours')
ax.legend()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60

"""
df = pd.read_sql_query(QUERY,conn)

med = df.groupby('Age')['ConvertedComp'].median()

med.plot(linestyle='-')
```

```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 25 AND 30

"""
df = pd.read_sql_query(QUERY,conn)

med = df.groupby('Age')['ConvertedComp'].median()

med.plot(linestyle='-')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY="""
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
"""
df= pd.read_sql_query(QUERY,conn)
print(df)
df.plot(kind='bar', color='yellow')
```

```python
QUERY="""
SELECT DevType, COUNT(*) as count
FROM DevType
GROUP BY DevType
ORDER BY Count DESC
"""
df= pd.read_sql_query(QUERY,conn)
print(df)
df.plot(kind='bar', color='orange')
```
